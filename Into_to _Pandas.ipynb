```python
#create weekly sales Data Frame
np.random.seed(0)
sales=np.random.randint(10,size=(5,3))
sales
```

```python
weekly_sales=pd.DataFrame(sales, index=["Monday","Tuesday","Wednesday","Thursday","Friday"], columns=["Peanut Butter","Almond Butter","Cashew Butter"])
```

```python
weekly_sales
```

```python
##prices Array
prices=np.array([10,8,12])
```

```python
 Butter_prices=pd.DataFrame(prices.reshape(1,3), index=["Prices"], columns=["Peanut Butter","Almond Butter","Cashew Butter"])
Butter_prices
```

```python
daily_sales_total=prices.dot(weekly_sales.T)
```

```python
daily_sales_total
```

```python
weekly_sales["Total ($)"]=daily_sales_total.T
```
